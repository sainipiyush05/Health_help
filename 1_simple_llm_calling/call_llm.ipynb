## Using Gemini API

```python
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI

load_dotenv()
```

```python
llm = ChatGoogleGenerativeAI(model="gemma-4-31b-it", temperature=0.9)

response   = llm.invoke("What is the capital of France?")
print(response.text)
```

```python
response = llm.invoke([
    ["system", "You are a helpful assistant that answers in short one word"],
    ["human", "What is the capital of France?"]
])
print(response.text)
```

```python
prompt = "Write a single tagline for a movie about a robot learning to love."

llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash", temperature=0)

response = llm.invoke(prompt)
print(response.text)
```

```python
prompt = "Write a single tagline for a movie about a robot learning to love."

llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash", temperature=1)

response = llm.invoke(prompt)
print(response.text)
```

## Using GROQ API

```python
from langchain_groq import ChatGroq
```

```python
llm = ChatGroq(model="qwen/qwen3-32b", reasoning_format="parsed")

message = [
    ["system", "You are a helpful assistant that answers in short one word"],
    ["human", "What is the capital of France?"]
]

ai_msg = llm.invoke(message)
print(ai_msg.text)
```
